# Zomato EDA

## Data loading



```python
import pandas as pd
df = pd.read_csv('zomato.csv', encoding='latin-1')
display(df.head())
print(df.shape)
```

## Data exploration

Explore the loaded data to understand its structure, identify missing values, data types, and basic statistics.


```python
display(df.head())
display(df.tail())

print(f"Shape of DataFrame: {df.shape}")

missing_values = df.isnull().sum()
print("\nMissing values per column:\n", missing_values)

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap')
plt.show()

print("\nData Types:\n", df.dtypes)

numerical_cols = df.select_dtypes(include=['number'])
print("\nDescriptive Statistics for Numerical Columns:\n", numerical_cols.describe())

categorical_cols = df.select_dtypes(exclude=['number'])
for col in categorical_cols.columns:
    print(f"\nUnique values and frequencies for {col}:\n{categorical_cols[col].value_counts()}")
```

## Data cleaning

Clean the data by handling missing values in the 'Cuisines' column and removing duplicate rows.


```python
most_frequent_cuisine = df['Cuisines'].mode()[0]
df['Cuisines'].fillna(most_frequent_cuisine, inplace=True)

df.drop_duplicates(inplace=True, keep='first')

display(df.head())
print(df.shape)
print(f"\nMissing values per column:\n{df.isnull().sum()}")
print(f"\nNumber of duplicates remaining: {df.duplicated().sum()}")
```

## Data analysis

Analyze the cleaned Zomato dataset to gain insights into restaurant ratings, cuisines, locations, and price ranges. Calculate key statistics and identify potential correlations between variables.


```python
import matplotlib.pyplot as plt
import seaborn as sns

rating_stats = df['Aggregate rating'].describe()
print("Rating Statistics:\n", rating_stats)

plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

cuisine_counts = df['Cuisines'].value_counts()

top_10_cuisines = cuisine_counts.head(10)

plt.figure(figsize=(12, 6))
top_10_cuisines.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

city_counts = df['City'].value_counts()

top_10_cities = city_counts.head(10)

plt.figure(figsize=(12, 6))
top_10_cities.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

average_rating_by_price = df.groupby('Price range')['Aggregate rating'].mean()
print(average_rating_by_price)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='viridis')
plt.title('Relationship between Price Range and Aggregate Rating')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt

average_cost_by_city = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False)

top_10_expensive_cities = average_cost_by_city.head(10)
bottom_10_expensive_cities = average_cost_by_city.tail(10)

plt.figure(figsize=(12, 6))
top_10_expensive_cities.plot(kind='bar', color='darkorange')
plt.title('Top 10 Cities with Highest Average Cost for Two')
plt.xlabel('City')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
bottom_10_expensive_cities.plot(kind='bar', color='lightgreen')
plt.title('Bottom 10 Cities with Lowest Average Cost for Two')
plt.xlabel('City')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df, color='purple')
plt.title('Average Cost vs. Aggregate Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

## Data visualization

Create visualizations to represent the findings from the data analysis stage. Visualize the distribution of ratings, the popularity of different cuisines, and the geographical distribution of restaurants.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], kde=True, color='skyblue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

cuisine_counts = df['Cuisines'].value_counts().head(10)
plt.figure(figsize=(12, 6))
cuisine_counts.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

city_counts = df['City'].value_counts().head(10)
plt.figure(figsize=(12, 6))
city_counts.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df, color='purple')
plt.title('Average Cost vs. Aggregate Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

## Observations

### Data Analysis Key Findings

1.Most restaurants have an aggregate rating above 3, indicating generally positive customer feedback.

2.North Indian cuisine is the most frequently listed cuisine.
There is a noticeable variation in the number of restaurants across different cities.

3.The average cost for two people varies across cities.
A higher average cost for two people does not guarantee a higher aggregate rating.

4.New Delhi has a high occurrence in the dataset, suggesting a significant presence in terms of the number of restaurants.

5.A substantial number of restaurants offer table booking services.
